```python
from sklearn.datasets import load_iris
iris=load_iris()
x=iris.data
y=iris.target
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=1,test_size=0.2)
```

```python
model=RandomForestClassifier()
model.fit(xtrain,ytrain)
model.score(xtest,ytest)
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
df=pd.read_csv(r"C:\Users\archa\OneDrive\Desktop\Downloads\ad.data",header=None)
df
```

```python
df.columns.values
```

```python
explanatory_variables=set(df.columns.values)
print(explanatory_variables)
```

```python
len(df.columns.values)
```

```python
response_variable=df[len(df.columns.values)-1]
response_variable
```

```python
explanatory_variables.remove(len(df.columns.values)-1)
print(explanatory_variables)
```

```python
x=df[list(explanatory_variables)]
x
```

```python
y=[1 if e=='ad.' else 0 for e in response_variable]
print(y)
```

```python
a=(x[0].unique())
print(a[5])
print(len(a[4]))
```

```python
x[x[0].values=='   ?']
```

```python
a=(x[2].unique())
print(a[6])
print(len(a[6]))
```

```python
x[x[2].values=='     ?']
```

```python
x.replace(to_replace=' *\?',value=-1,regex=True,inplace=True) #regular expression, regex=True:- to understand it is regular expression
x
```

```python
a=(x[2].unique())
print(a)
print(a[6])
```

```python
import numpy as np
y=np.array(y)
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.5)
print(xtrain.shape)
print(ytrain.shape)
```

```python
pipeline=Pipeline([('clf',DecisionTreeClassifier(criterion='entropy'))])
parameter={
    'clf__max_depth':(100,150,155,160),
    'clf__min_samples_leaf':(1,2,3,6,8)
}
grid_search=GridSearchCV(pipeline,parameter,n_jobs=-1,verbose=1,scoring='f1')
grid_search.fit(xtrain,ytrain)
```

```python
print('score',grid_search.best_score_)
```

```python
best_parameter=grid_search.best_estimator_.get_params()
best_parameter
```

```python
best_max_depth=best_parameter['clf__max_depth']
best_min_leaf=best_parameter['clf__min_samples_leaf']
best_max_depth
```

```python
best_min_leaf
```

```python
model=DecisionTreeClassifier(max_depth=best_max_depth,min_samples_leaf=best_min_leaf)
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
from sklearn.ensemble import RandomForestClassifier
pipeline=Pipeline([('clf',RandomForestClassifier(criterion='entropy'))])
parameters={
    'clf__max_depth':(100,150,155,160),
    'clf__min_samples_leaf':(1,2,3,6,8),
    'clf__min_samples_split':(2,3),
    'clf__n_estimators':(50,70,100)
}
```

```python
grid_search=GridSearchCV(pipeline,parameters,n_jobs=-1,scoring='f1')
grid_search.fit(xtrain,ytrain)
```

```python
print('score',grid_search.best_score_)
```

```python
best_parameter=grid_search.best_estimator_.get_params()
best_parameter
```

```python
best_max_depth=best_parameter['clf__max_depth']
best_min_leaf=best_parameter['clf__min_samples_leaf']
best_min_samples_split=best_parameter['clf__min_samples_split']
best_n_estimator=best_parameter['clf__n_estimators']
best_max_depth
```

```python
best_min_leaf
```

```python
best_min_samples_split
```

```python
best_n_estimator
```

```python
model=RandomForestClassifier(max_depth=100,min_samples_leaf=1,min_samples_split=3,n_estimators=170)
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```
